```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("customers.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.drop(columns=["customerID", "Churn"], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = df.select_dtypes(include="object").columns

for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(df)
```

```python
print(X_scaled[:5])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
```

```python
print(X_pca.shape)
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
print("Total Variance Explained:", pca.explained_variance_ratio_.sum())
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
for i, value in enumerate(wcss, start=2):
    print(f"Clusters = {i}, WCSS = {value:.2f}")
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_pca)

    score = silhouette_score(X_pca, labels)
    silhouette_scores.append(score)

    print(f"Clusters = {i}, Silhouette Score = {score:.4f}")
```

```python
best_k = silhouette_scores.index(max(silhouette_scores)) + 2

print("Best Number of Clusters:", best_k)
print("Highest Silhouette Score:", max(silhouette_scores))
```

```python
best_k = 3

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)

clusters = kmeans.fit_predict(X_pca)
```

```python
df["Cluster"] = clusters
```
